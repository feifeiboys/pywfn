## 此脚本用来分析错误原因
1. err-1: s+d obtial contribute is biger
2. err-2: sp contribute is smaller
3. err-3: p contribute is smaller
4. err-4: s+d contribute is bigger than s+p contribute
5. err-5: there are cross profile in p obtial way
6. err-6: p obtial vector is not verpendicular to bond
7. err-7:around normal and p orbital angle is far away
8. err-8: not increase or decrease at the same time
9. err-9: all angle is too big
10. err-10: center atom connection number not equal to 2 or 3

```python
import os
import re
import matplotlib.pyplot as plt
```

```python
def analyze(idx):
    logs=os.listdir('logs')
    with open(f'logs/{logs[idx]}','r') as f:
        content=f.read()
    nums=[]
    for i in range(10):
        num=len(re.findall(f'err-{i+1}',content))
        nums+=[num]
    plt.pie(nums,labels=[f'err-{i+1}' if nums[i]!=0 else '' for i in range(10)])
    plt.show()
```

```python
analyze(-1),analyze(-2),analyze(-3)
```

```python
analyze(-2)
```
