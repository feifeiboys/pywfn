```python
import os
import re
import matplotlib.pyplot as plt
```

```python
def analyze(idx):
    errLength=12
    logs=os.listdir('logs')
    with open(f'logs/{logs[idx]}','r') as f:
        content=f.read()
    nums=[]
    for i in range(errLength):
        num=len(re.findall(f'err-{i+1}',content))
        nums+=[num]
    # plt.pie(nums,labels=[f'err-{i+1}' if nums[i]!=0 else '' for i in range(errLength)])
    plt.bar([f'err-{i+1}' if nums[i]!=0 else '' for i in range(errLength)],nums)
    plt.show()
```

```python
analyze(-1)
```
