```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
orders1=np.load('orders1.npy')
orders2=np.load('orders2.npy')
orders3=np.load('orders3.npy')
```

```python
angles=[(i*10)+90 if i<=27 else i*10-270 for i in range(37)]
angles=[f'{each:.0f}' for each in angles]
```

```python
step=4
indexs=list(range(0,len(orders1),step))
labels=[angles[i] for i in indexs]
plt.xticks(indexs,labels,rotation =35)
plt.title(f'Bond Order - Dihedral Angle')
plt.xlabel('Dihedral Angle')
plt.ylabel('Bond Order')
plt.plot(np.arange(len(orders1)),orders1,'-*')
plt.savefig('orders1.png',dpi=300, bbox_inches = 'tight')
```
