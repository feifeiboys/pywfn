```python
import numpy as np
import pandas as pd
import re
```

```python
# 轨道系数 C
CM=pd.read_csv('DATA.csv').to_numpy()
pd.DataFrame(CM)
```

```python
np.sum(CM[0,:8]**2)*2
```

```python
# 密度矩阵 P(自己计算的)
P=np.zeros((10,10))
n=np.array([2]*8+[0]*2)
for i in range(10):
    C1=CM[:,i][:,np.newaxis]
    C2=CM[:,i][np.newaxis,:]
    P+=C1*C2*n[i]
pd.DataFrame(P)
```

```python
# 重叠矩阵 S
overlapStr=\
"""0.100000D+01
0.236704D+00  0.100000D+01
0.000000D+00  0.000000D+00  0.100000D+01
0.000000D+00  0.000000D+00  0.000000D+00  0.100000D+01
0.000000D+00  0.000000D+00  0.000000D+00  0.000000D+00  0.100000D+01
0.000000D+00  0.222396D-01  0.000000D+00  0.000000D+00 -0.376360D-01 0.100000D+01
0.222396D-01  0.276700D+00  0.000000D+00  0.000000D+00 -0.317826D+00 0.236704D+00  0.100000D+01
0.000000D+00  0.000000D+00  0.149257D+00  0.000000D+00  0.000000D+00 0.000000D+00  0.000000D+00  0.100000D+01
0.000000D+00  0.000000D+00  0.000000D+00  0.149257D+00  0.000000D+00 0.000000D+00  0.000000D+00  0.000000D+00  0.100000D+01
0.376360D-01  0.317826D+00  0.000000D+00  0.000000D+00 -0.305811D+00 0.000000D+00  0.000000D+00  0.000000D+00  0.000000D+00  0.100000D+01"""
overList=[]
for line in overlapStr.split('\n'):
    data=re.split(' +',line)
    data=[float(each.replace('D','e')) for each in data]
    data+=[0.]*(10-len(data))
    overList.append(data)
overMatrix=np.array(overList)
S=np.tril(overMatrix)+np.tril(overMatrix,-1).T
pd.DataFrame(S)
```

```python
P12=P[0:5,5:10]
P21=P[5:10,0:5]
S12=S[0:5,5:10]
S21=S[5:10,0:5]
PS=P@S
```
