```python
def progress(idx,total):
    """打印进度条;idx:当前的进度;total:总任务数"""
    percent=idx/total
    num=int(percent*10)
    end='\n' if percent==1 else ''
    print(f'\r{"*"*num}{"_"*(10-num)}{idx}/{total}',end=end)
```

```python
import time
for i in range(100):
    progress(i+1,100)
    time.sleep(0.1)
```
